```python
import numpy as np

# Descriptive Statistics
print("Mean:\n", retail_df[["Price", "Quantity", "Discount"]].mean())
print("\nMedian:\n", retail_df[["Price", "Quantity", "Discount"]].median())
print("\nMode:\n", retail_df[["Price", "Quantity", "Discount"]].mode().iloc[0])
print("\nStandard Deviation:\n", retail_df[["Price", "Quantity", "Discount"]].std())
print("\nVariance:\n", retail_df[["Price", "Quantity", "Discount"]].var())
print("\nMin:\n", retail_df[["Price", "Quantity", "Discount"]].min())
print("\nMax:\n", retail_df[["Price", "Quantity", "Discount"]].max())
print("\nQuantiles:\n", retail_df[["Price", "Quantity", "Discount"]].quantile([0.25, 0.5, 0.75]))
print("\nDescribe():\n", retail_df[["Price", "Quantity", "Discount"]].describe())

# Trigonometric Functions (on Quantity column)
print("\nSin:\n", np.sin(retail_df["Quantity"]))
print("\nCos:\n", np.cos(retail_df["Quantity"]))
print("\nTan:\n", np.tan(retail_df["Quantity"]))

# Logarithmic & Exponential Functions (on Price column)
print("\nNatural Log (ln):\n", np.log(retail_df["Price"]))
print("\nLog base 10:\n", np.log10(retail_df["Price"]))
print("\nLog base 2:\n", np.log2(retail_df["Price"]))
print("\nExponential (e^x for Quantity):\n", np.exp(retail_df["Quantity"]))

```

```python
# === Trigonometric Functions (on Quantity column) ===
print("\nSin:\n", retail_df["Quantity"].apply("sin"))
print("\nCos:\n", retail_df["Quantity"].apply("cos"))
print("\nTan:\n", retail_df["Quantity"].apply("tan"))

# === Logarithmic & Exponential Functions (on Price column) ===
print("\nNatural Log (ln):\n", retail_df["Price"].apply("log"))
print("\nLog base 10:\n", retail_df["Price"].apply("log10"))
print("\nLog base 2:\n", retail_df["Price"].apply("log2"))
print("\nExponential (e^x for Quantity):\n", retail_df["Quantity"].apply("exp"))
```
