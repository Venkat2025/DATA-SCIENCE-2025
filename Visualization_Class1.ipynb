SALES TREND OVER TIME

```python
df.columns
```

```python
sales_by_date = df.groupby('Order Date')['Sales'].sum()
```

```python
sales_by_date.head()
```

```python
sales_by_date.index[0:5]
```

```python
sales_by_date.values
```

```python
plt.figure(figsize=(20,10))
plt.plot(sales_by_date.index,sales_by_date.values)
plt.title("Daily Sales Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.dates as mdates

plt.figure(figsize=(20,10))
plt.plot(sales_by_date.index, sales_by_date.values)
plt.title("Daily Sales Trend")

# Auto-format dates
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.plot(sales_by_date.index, sales_by_date.values,color='y',linestyle='--')
plt.title("Daily Sales Trend",fontsize=24)
plt.xlabel("Order Dates",fontsize=24)
plt.ylabel("Sales",fontsize=24)
plt.grid(True,linestyle='--')

# show only every 30th tick
plt.xticks(sales_by_date.index[::30], rotation=45)

plt.show()

```

Daily Profit Trend


```python
import matplotlib.dates as mdates

plt.figure(figsize=(20,10))
plt.plot(sales_by_date.index, sales_by_date.values,color='g',linestyle='--')
plt.title("Daily Profit Trend",fontsize=24)
plt.xlabel("Order Dates",fontsize=24)
plt.ylabel("Profit",fontsize=24)
plt.grid(True,linestyle='--')

# Auto-format dates
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
sales_by_date = df.groupby('Order Date')['Profit'].sum()

plt.figure(figsize=(20,10))
plt.plot(sales_by_date.index, sales_by_date.values,color='r',linestyle='--')
plt.title("Daily Profit Trend",fontsize=24)
plt.xlabel("Order Dates",fontsize=24)
plt.ylabel("Profit",fontsize=24)
plt.grid(True,linestyle='--')
# show only every 30th tick
plt.xticks(sales_by_date.index[::30], rotation=45)

plt.show()
```

Region Wise Sales

```python
import matplotlib.dates as mdates

plt.figure(figsize=(20,15))
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.bar(sales_by_region.index, sales_by_region.values,color='r')
plt.title("Sales By Region Trend",fontsize=24)
plt.xlabel("Region",fontsize=24)
plt.ylabel("Total Sales",fontsize=24)
#plt.grid(True)

# Auto-format dates
# plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
# plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,15)) 
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.barh(sales_by_region.index, sales_by_region.values,color=['#F54927','#0EED6B','#F52C6C','#2CE1F5'])
plt.title("Sales By Region Trend",fontsize=24)
plt.xlabel("Region",fontsize=24)
plt.ylabel("Total Sales",fontsize=24)
plt.xticks(rotation=90)
plt.show()
```

```python
# STACKED BAR GRAPH
#lt.figure(figsize=(20,15))
sales_by_region = df.groupby('Region')['Sales'].sum()
profit_by_region = df.groupby('Region')['Profit'].sum()
plt.bar(sales_by_region.index, sales_by_region.values,color='r')
plt.bar(sales_by_region.index, profit_by_region.values,bottom=sales_by_region.values,color='b',label='Profit')
plt.title("Sales By Region Trend",fontsize=24)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Total Sales",fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# STACKED BAR GRAPH
#lt.figure(figsize=(20,15))
sales_by_region = df.groupby('Region')['Sales'].sum()
profit_by_region = df.groupby('Region')['Profit'].sum()
plt.bar(sales_by_region.index, sales_by_region.values,facecolor='w',edgecolor='y')
plt.bar(sales_by_region.index, profit_by_region.values,bottom=sales_by_region.values,color='b',label='Profit',facecolor='g',edgecolor='r')
plt.title("Sales By Region Trend",fontsize=24)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Total Sales",fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
